```python
User_name = input("Enter your name: ")
Daily_budget = float(input("Enter your daily budget: "))

Food_expenses = float(input("Enter your expense for food: "))
Transport_expenses = float(input("Enter your expense for transport: "))
Miscellaneous_expenses = float(input("Enter your expense for others: "))

Expenses = (Food_expenses, Transport_expenses, Miscellaneous_expenses)

Total_expenses = Expenses[0] + Expenses[1] + Expenses[2]

if Total_expenses > Daily_budget:
    print(f"You have exceeded your budget by {Total_expenses - Daily_budget:}.")
else:
    print(f"You are within your budget with {Daily_budget - Total_expenses:} left.")

print(f"Total expenses: {Total_expenses:}")
```
